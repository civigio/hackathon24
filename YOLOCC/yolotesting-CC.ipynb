```python
probs = {}
for imgname in imgnames:
    img = mpimg.imread(imgname)
    index = int(imgname[-7:-4])
    labelfile = os.path.join(r"C:\Users\civig\Desktop\hackathon24\YOLOCC\runs\classify\predict8\labels", f"frame{index}.txt")
    with open(labelfile, "r") as f:
        lines = f.readlines()
        #print(lines)
        outcomes = {}
        for line in lines:
            # read number and string in line
            number, string = line.split(" ", 1)
            #print(number, string)
            outcomes[string] = float(number)
        #print(index, outcomes)
        probs[index] = outcomes
```
